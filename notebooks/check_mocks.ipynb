# Measure the power spectrum from a single mock

This computes one auto or cross pseudo power spectrum, and undoes the mode coupling from the mask, using Namaster. 
It loads the theory prediction for the corresponding power spectrum, applies the mask mode coupling to it, bins it, then deconvolves the mask mode coupling, so that it can be compared to the measurement.
